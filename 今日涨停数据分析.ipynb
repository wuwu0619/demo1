```python
import pandas as pd 
import numpy as np      
import matplotlib.pyplot as plt
from datetime import datetime
plt.rcParams['font.sans-serif']=['SimHei'] 
plt.rcParams['axes.unicode_minus']=False
```

```python
df=pd.read_excel(r"C:\Users\Admin\Desktop\影刀数据表格_20250314-150344.xlsx")
df
```

```python
df = df.iloc[1:-1,1:-1]
df
```

```python
df.info()
```

```python
def convert_units(value):
    try:
        num, unit = float(value[:-1]), value[-1]
        if unit == '亿':
            return num * 100000000
        elif unit == '万':
            return num * 10000
        else:
            return num  # 如果没有识别出有效的单位，默认返回原数字
    except Exception:
        return value  # 对于无法解析的情况保持原有值不变
```

```python
df.loc[:,'成交额'] = df['成交额'].apply(convert_units)
df.loc[:, '流通市值'] = df['流通市值'].apply(convert_units)
df.loc[:, '总市值'] = df['总市值'].apply(convert_units)
df.loc[:, '封板资金'] = df['封板资金'].apply(convert_units)
```

```python
df
```

```python
df['涨跌幅'] = df['涨跌幅'].str.replace('%', '').astype(float) / 100
df['换手率'] = df['换手率'].str.replace('%', '').astype(float) / 100
df['炸板次数'] = df['炸板次数'].str.replace('次', '').astype(int)
```

```python
df.info()
```

```python
df['成交额'] = df['成交额'].astype(np.int64)  
df['流通市值'] = df['流通市值'].astype(np.int64) 
df['总市值'] = df['总市值'].astype(np.int64)
df['封板资金'] = df['封板资金'].astype(np.int64)
df['首次封板时间'] = pd.to_datetime(df['首次封板时间'])
df['最后封板时间'] = pd.to_datetime(df['最后封板时间'])
```

```python
industry_counts = df['所属行业'].value_counts().head(10)  # 仅此一行改动
plt.figure(figsize=(10, 6))

colors = plt.cm.tab20(np.arange(len(industry_counts)))  # 自动根据前10个行业生成颜色

ax = industry_counts.plot(
    kind='bar',
    color=colors,
    edgecolor='black',
    linewidth=0.7,
    width=0.8
)

plt.title('各行业涨停股票数量TOP10', fontsize=14, pad=20)
plt.xlabel('行业', fontsize=12, labelpad=10)
plt.ylabel('数量', fontsize=12, labelpad=10)
plt.xticks(rotation=45, ha='right')

ax.yaxis.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.savefig('各行业涨停股票数量TOP10.png')
plt.show()
```

```python
# 计算各行业平均封板资金并按降序排列
industry_avg = df.groupby('所属行业')['封板资金'].mean().sort_values(ascending=False).head(10)

plt.figure(figsize=(14, 8))

# 使用渐变色+标记点样式
ax = industry_avg.plot(
    kind='line',
    marker='o',                # 圆形数据点
    markersize=8,              # 点大小
    linewidth=2.5,             # 线宽
    color=plt.cm.tab20(0),     # 使用tab20首色
    markerfacecolor='white',   # 标记点填充色
    markeredgewidth=2          # 标记点边框粗细
)

# 样式增强
plt.title('涨停股票各行业平均封板资金TOP10', 
         fontsize=16, 
         pad=20, 
         fontweight='bold',
         color='#2c3e50')

plt.xlabel('行业', 
          fontsize=12, 
          labelpad=12,
          color='#34495e')

plt.ylabel('平均封板资金 (亿元)', 
          fontsize=12,
          labelpad=12,
          color='#34495e')

# 坐标轴优化
plt.xticks(
    range(len(industry_avg)),  # 设置x轴刻度位置
    industry_avg.index,        # 设置刻度标签
    rotation=45, 
    ha='right', 
    fontsize=11
)

# 添加数据标签
for x, y in enumerate(industry_avg):
    ax.text(x, y*1.02,         # 上浮2%防止重叠
           f'{y/100000000:.1f}亿',
           ha='center', 
           va='bottom',
           fontsize=10,
           color='#2c3e50')

# 网格线优化
ax.yaxis.grid(True, 
             linestyle='--', 
             alpha=0.6,
             color='#7f8c8d')

# 移除边框
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

# 设置背景色
ax.set_facecolor('#f8f9fa')

# 调整y轴范围
plt.ylim(industry_avg.min()*0.9, industry_avg.max()*1.1)

plt.tight_layout()
plt.savefig('涨停股票各行业平均封板资金TOP10.png')
plt.show()
```

```python
top10_broken = df.sort_values(by='炸板次数', ascending=False).head(10)
# 绘制炸板次数TOP10的折线图
plt.figure(figsize=(10, 6))
plt.plot(top10_broken['名称'], top10_broken['炸板次数'], marker='o', linestyle='-', color='red')
plt.title('今日涨停股票炸板次数TOP10')
plt.xlabel('股票名称')
plt.ylabel('炸板次数')
plt.xticks(rotation=45)
plt.grid(True)
plt.savefig('今日涨停股票炸板次数TOP10.png')
plt.show()
```

```python
# 按成交额排序并取前10
top10_turnover = df.sort_values(by='成交额', ascending=False).head(10)

plt.figure(figsize=(12, 7))  # 调整为更合适的画布尺寸

# 使用现代配色方案（示例使用tab10颜色映射）
colors = plt.cm.tab10(np.arange(len(top10_turnover)))

ax = plt.bar(
    top10_turnover['名称'], 
    top10_turnover['成交额'],
    color=colors,
    edgecolor='white',  # 白色边框
    linewidth=1.5,     # 边框粗细
    width=0.8          # 柱宽调整
)

# 样式增强
plt.title('今日涨停股票成交额TOP10', 
          fontsize=15, pad=20, 
          fontweight='bold', 
          color='#2c3e50')  # 深灰色标题

plt.xlabel('股票名称', 
           fontsize=12, 
           labelpad=12, 
           color='#34495e')  # 协调的字体颜色

plt.ylabel('成交额（亿元）',  # 建议添加单位
          fontsize=12,
          labelpad=12,
          color='#34495e')

plt.xticks(rotation=45, 
           ha='right',  # 标签右对齐
           fontsize=11)

# 添加数据标签
for bar in ax:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2., 
             height + 0.02*height,  # 在柱顶上方5%高度处显示
             f'{height/100000000:.1f}亿',  # 格式化显示单位（假设成交额单位为元）
             ha='center', 
             va='bottom',
             fontsize=10,
             color='#2c3e50')

# 网格线优化
plt.grid(axis='y', 
         linestyle='--', 
         alpha=0.6, 
         color='#7f8c8d')

# 移除顶部和右侧边框
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)

# 设置背景色
plt.gca().set_facecolor('#f8f9fa')

plt.tight_layout()
plt.savefig('今日涨停股票成交额TOP10.png')
plt.show()
```

```python
# 获取封板资金前十股票数据
top10_capital = df.sort_values(by='封板资金', ascending=False).head(10)

plt.figure(figsize=(12, 7))

# 创建渐变色方案（使用Spectral色系）
colors = plt.cm.Spectral(np.linspace(0.1, 0.9, len(top10_capital)))

ax = plt.bar(
    top10_capital['名称'],
    top10_capital['封板资金'],
    color=colors,
    edgecolor='white',  # 白色边框
    linewidth=1.5,
    width=0.75
)

# 图表样式配置
plt.title('今日涨停股票封板资金TOP10', 
         fontsize=16, 
         pad=20, 
         fontweight='bold',
         color='#2c3e50')  # 深灰色标题

plt.xlabel('股票名称', 
          fontsize=12, 
          labelpad=12,
          color='#34495e')  # 协调的字体颜色

plt.ylabel('封板资金（亿元）', 
          fontsize=12,
          labelpad=12,
          color='#34495e')

# X轴标签优化
plt.xticks(rotation=45, 
          ha='right',  # 标签右对齐
          fontsize=11)

# 添加数据标签（带单位格式化）
for bar in ax:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2., 
             height * 1.005,  # 上浮0.5%
             f'{height/100000000:.1f}亿',  # 假设原始单位为元
             ha='center', 
             va='bottom',
             fontsize=10,
             color='#2c3e50')

# 网格线优化
plt.grid(axis='y', 
        linestyle=':', 
        alpha=0.7,
        color='#95a5a6')

# 边框优化
ax = plt.gca()
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['left'].set_color('#bdc3c7')

# 背景色设置
ax.set_facecolor('#f5f6fa')

# 自动调整Y轴范围
max_value = top10_capital['封板资金'].max()
plt.ylim(0, max_value * 1.15)  # 留出15%空白

plt.tight_layout()
plt.savefig('今日涨停股票封板资金TOP10.png')
plt.show()
```
